```python
# q.48 bubble sort
numbers = [5, 3, 8, 1]

n = len(numbers)

for i in range(n):
    for j in range(n - i - 1):
        if numbers[j] > numbers[j + 1]:
            numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]

print(numbers)
```

```python
# q.49 selection sort
numbers = [64, 25, 12, 22, 11]

n = len(numbers)

for i in range(n):

    min_index = i

    for j in range(i + 1, n):
        if numbers[j] < numbers[min_index]:
            min_index = j

    numbers[i], numbers[min_index] = numbers[min_index], numbers[i]

print(numbers)
```

```python
# q.51 binary search
numbers = [10, 20, 30, 40, 50, 60, 70]

key = int(input("Enter number to search: "))

low = 0
high = len(numbers) - 1

while low <= high:

    mid = (low + high) // 2

    if numbers[mid] == key:
        print("Element found at index", mid)
        break

    elif numbers[mid] < key:
        low = mid + 1

    else:
        high = mid - 1

else:
    print("Element not found")
```

```python
# q.77 recursive factorial
def fact(number):
    if number == 0 or number == 1:
        return 1
    else:
        return number * fact(number - 1)
    
number = int(input("enter a number to find factorial : "))
print(fact(number))
```

```python
# q.78 recursive fibonacci
def fibonacci(number):
    if number == 0:
        return 0
    elif number == 1:
        return 1
    else:
        return fibonacci(number - 1) + fibonacci(number - 2)

number = int(input("enter a number to find fibonacci : "))
print(fibonacci(number))

for i in range(number):
    print(fibonacci(i), end=" ")
```

```python
# q.86 polymorphism

# method overriding ---> runtime polymorphism
class A:
    def show(self):
        print("A")
class B(A):
    def show(self):
        print("B")
class C(A):
    def show(self):
        print("C")
class D(B, C):
    pass

d = D()
d.show()

# built in method polymorphism, operator overloading
print(len("Python"), len([1,2,3,4,5]), len((1,2,3)), len({1 : 10, 2 : 20}))
print((10 + 2), ([1, 2] + [3, 4]), ("Khushi" + "\t" + "Agrawal"), ((5, 6) + (7, 8)))

# method overloading
class A:
    def show(self, b = None):
        if b == None:
            print("None")
        else:
            print(b)
a = A()
a.show()
a.show("Khushi Agrawal")
```

```python
# q.90 Duck Typing
# if it walks like a duck and quacks like a duck, then treat it as a duck
class ac_remote:
    def operation(self):
        print("ac remote is operating")
class tv_remote:
    def operation(self):
        print("tv remote is operating")
def remote(kind):
    kind.operation()


ac, tv = ac_remote(), tv_remote()
remote(ac)
remote(tv)
```
